```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import json
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import cm
from datetime import datetime
import glob
import seaborn as sns
```

```python
files = [i for i in glob.glob('data/*.csv')]
sorted(files)
```

```python
dfs = list()
for file in files:
    print(file)
    country = ''.join([c for c in file if c.isupper()])
    df = pd.read_csv(file)
    df['country'] = country
    dfs.append(df)
    
df = pd.concat(dfs)
```

```python
df
```

```python
df.info()
```

```python
df = df.dropna(how='any',inplace=False, axis = 0)
```

```python
df.info()
```

```python
df = df.drop_duplicates(keep="last")
```

```python
df.info()
```

```python
category_id = {}

with open('data/US_category_id.json', 'r') as f:
    data = json.load(f)
    for category in data['items']:
        category_id[category['id']] = category['snippet']['title']
df['category_id']=df['category_id'].astype(str).map(category_id) 
print(df['category_id'].unique())
```

```python
df_likes = df[df['country']=='IN'][['likes','category_id']] \
            .groupby(by='category_id') \
            .mean() \
            .sort_values(by=['likes'])
plt.figure(figsize=(14,9))
ax = sns.barplot(y=df_likes.index,x=df_likes['likes'],orient='h')
plt.xlabel("M√©dia de Likes") 
plt.ylabel("Categorias")
```

```python
labels = df.groupby(['country']).count().index
sizes = df.groupby(['country']).count()['title']

fig, ax = plt.subplots(figsize=(10,10))
ax.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True)
ax.axis('equal') 
```
